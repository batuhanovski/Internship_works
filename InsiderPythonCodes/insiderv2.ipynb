```python
def todaynumber(date):
    if date[6:7] == '1':
        return int(date[8:10])  
    elif date[6:7] == '2':
        return int(date[8:10])+31
    elif date [6:7] == '3':
        return int(date[8:10])+60
```

```python
cmpocs['daynum']=cmpocs.apply(lambda cmpocs: todaynumber(cmpocs['date']),axis=1)
```

#  Create points for every user

```python
#different from number of session creations by a user, this point includes the value of the date session has created
```

```python
rstcmpocs=cmpocs.reset_index()
```

```python
def pointation(day):
    if day<8:
        return 1
    elif day<16 and day>7:
        return 1.2
    elif day<24 and day>15:
        return 1.4
    elif day<32 and day>23:
        return 1.6
    elif day<42 and day>32:
        return 1.8
    elif day<50 and day>41:
        return 2
    else:
        return 2.2
```

```python
rstcmpocs['points']=rstcmpocs.apply(lambda rstcmpocs: pointation(rstcmpocs['daynum']),axis=1)
```
